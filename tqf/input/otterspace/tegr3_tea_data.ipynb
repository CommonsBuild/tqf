```python
import pandas as pd
import json
import hvplot.pandas
```

```python
with open('tegr3_otterspace_1.json', 'r') as f:
    data = json.load(f)
tea_1 = pd.DataFrame(data['results'])
tea_1
```

```python
with open('tegr3_otterspace_2.json', 'r') as f:
    data = json.load(f)
tea_2 = pd.DataFrame(data['results'])
tea_2
```

```python
with open('tegr3_otterspace_3.json', 'r') as f:
    data = json.load(f)
tea_3 = pd.DataFrame(data['results'])
tea_3
```

```python
with open('tegr3_otterspace_4.json', 'r') as f:
    data = json.load(f)
tea_4 = pd.DataFrame(data['results'])
tea_4
```

```python
tea = pd.concat([tea_1, tea_2, tea_3, tea_4])
tea
```

```python
tea['id'].nunique()
```

```python
tea.isna().sum()
```

```python
tea[tea.duplicated('id')].groupby('id')['totalBadgesInCommunityCount'].apply(list)
```

```python
tea = tea.drop_duplicates(subset=['id']).reset_index(drop=True)
```

```python
tea.hvplot.scatter(y='totalBadgesInCommunityCount', x='index', title="TEA Credentials Balances Scatter Plot", alpha=0.8)
```

```python
tea.sort_values('totalBadgesInCommunityCount', ascending=False).reset_index(drop=True).hvplot.line(y='totalBadgesInCommunityCount', x='index', title='TEGR2 TEA Credentials')
```

```python
tea.groupby('totalBadgesInCommunityCount').count().hvplot.bar(y='id', title="TEA Credentials Balances Bar Chart", ylabel="Wallet Count", alpha=0.8)
```

```python
tea.to_csv('tegr3_tea_holders.csv', index=False)
```
